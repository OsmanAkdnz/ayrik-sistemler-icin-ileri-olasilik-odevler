```python
from scipy.stats import poisson
import numpy as np 
import matplotlib.pyplot as plt

samples= np.arange(0,100,0.1)
pmf = poisson.pmf(samples,50)
print(np.sum(pmf))  # toplamı 1 e eşit 
plt.plot(samples, pmf)
plt.show()

```
