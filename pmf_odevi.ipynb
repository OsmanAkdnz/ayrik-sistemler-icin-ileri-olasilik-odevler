0 ile 9 arasında 100 tane seçim yapılıyor. PMF dağılımı grafiğini gösteriniz

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

 
xs = np.random.randint(0, 10, 100)
print(xs)
plt.ylabel('PMF')
sns.histplot(xs, stat='probability', bins=50);
```
