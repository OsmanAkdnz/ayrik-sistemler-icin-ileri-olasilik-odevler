Bir ağacın yuz yılda 0.2 aralıkta olacak şekilde 0 ile 100 cm aralığında boyu uzuyor. Bu ağacın 1 sonraki yılda uzama miktarının olasılıksal dağılımı nedir?


samples= np.arange(0,100,0.5)   0 ile 100 arasında 0.2 artan sample oluşturuyoruz.
pmf = poisson.pmf(samples,mean) ortalama değer için olasılıklarını oluşturuyoruz
plt.plot(samples, pmf) grafiğe dökuyoruz.

```python
from scipy.stats import poisson
import numpy as np 
import matplotlib.pyplot as plt

samples= np.arange(0,100,0.2)
mean = np.sum(samples) / len(samples)
pmf = poisson.pmf(samples,mean)
print(np.sum(pmf))  # toplamı 1 e eşit 
plt.plot(samples, pmf)
plt.show()

```
