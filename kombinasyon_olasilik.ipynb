10 kiz 6 erkek olan bir sinifta secilecek olan 3 kisinin en az ikisinin erkek olma olasiligi nedir?

secimler = eek veya eee ise 

Cozum = (C(6,2) * C(10,1) + C(6,3)) / C(16,3) = 0.30357142857142855



```python
def calculate_factoriel(n) :
   if n < 2:
        return 1
   else:
        return n * calculate_factoriel(n-1)

def calc_comb(element_num,selected_num):
    n = calculate_factoriel(element_num)
    r = calculate_factoriel(selected_num)
    n_r = calculate_factoriel(element_num - selected_num)
    return n / (r * n_r)

olasilik = (calc_comb(6,2) * calc_comb(10,1) + calc_comb(6,3)) / calc_comb(16,3)  
print(f"olasilik = {olasilik}")
```
