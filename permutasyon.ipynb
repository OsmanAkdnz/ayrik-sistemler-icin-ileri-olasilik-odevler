6 kisinin yan yana fotografi cekilecektir. Bu 6 kisi yanyana kac farkli sekilde fotograf cekilebilir


sonuc = P(6,6) = 6! / (6-6)! = 720

```python
def calculate_factoriel(n) :
   if n < 2:
        return 1
   else:
        return n * calculate_factoriel(n-1)

def calc_perm(n,r):
    n_fac = calculate_factoriel(n)
    n_r = calculate_factoriel(n - r)
    return n_fac /n_r


result = calc_perm(6,6)
print(f"sonuc = {result}")
```
