Bir torbada 4 u kirmizi 6 si sari renki 10 bilye vardir bu torbadan alinan 2 bilyenin birinin kirmizi renkli oldugu bilindigine gore sari renkli olma olasiligi nedir?

P(A|B) = P(Aâˆ©B) / P(B)
Olasilik = Birinin kirmizi,digerinin sari gelme olasiligi / Birinin kirmizi gelme olasiligi 

olasilik = (C(4,1)*C(6,1) / C(10,2)) / 1 - C(6,2) / C(10,2) (sari gelmeme olasiligi)

olasililik = 4/5

```python
def calculate_factoriel(n) :
   if n < 2:
        return 1
   else:
        return n * calculate_factoriel(n-1)

def calc_comb(element_num,selected_num):
    n = calculate_factoriel(element_num)
    r = calculate_factoriel(selected_num)
    n_r = calculate_factoriel(element_num - selected_num)
    return n / (r * n_r)
    
olasilik = ((calc_comb(4,1) * calc_comb(6,1)) / calc_comb(10,2)) / (1 - (calc_comb(6,2) / calc_comb(10,2)))

print(f"olasilik = {olasilik}")
```
